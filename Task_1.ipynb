# **TASK 1**
Implement a simple quantum operation with Cirq or Pennylane

1.   With 5 qubits
2.   Apply Hadamard operation on every qubit
3.   Apply CNOT operation on (0, 1), (1,2), (2,3), (3,4)
4.   SWAP (0, 4)
5.   Rotate X with pi/2 on any qubit
6.   Plot the circuit









```python
#importing all the libraries
from functools import partial
import pennylane as qml
import matplotlib.pyplot as plt
import math

#Setting up the device for task 1
dev = qml.device("default.qubit", wires=5)
```

```python
#scripting the function to construct the circuit
@qml.qnode(dev)
def qn():
  for i in range(5):
    qml.Hadamard(wires = i)
  for j in range(4):
    qml.CNOT(wires = [j,j+1])
  qml.SWAP(wires = [0,4])
  qml.RX(math.pi/2,wires=2)
  return qml.state()
```

```python
#Drawing the circuit
fig,ax = qml.draw_mpl(qn)()
```

# **TASK 2**
Implement a second circuit with a framework of your choice:


1.   Apply a Hadmard gate to the first qubit
2.   rotate the second qubit by pi/3 around X
3.   Apply Hadamard gate to the third and fourth qubit
4.   Perform a swap test between the states of the first and second qubit |5.   q1 q2> and the third and fourth qubit |q3 q4>


```python
#Setting up the device for task 2
dev_2 = qml.device("default.qubit", wires=6)

#scripting the function to construct the circuit
@qml.qnode(dev_2)
def qn_2():
  qml.Hadamard(wires = 0)
  qml.RX(math.pi/3,wires=1)
  qml.Hadamard(wires = 2)
  qml.Hadamard(wires = 3)

  #Swap Test between states of first and second qubit
  qml.Hadamard(wires = 4)
  qml.CSWAP(wires = [4,0,1])
  qml.Hadamard(wires = 4)

  #Swap Test between states of third and fourth qubit
  qml.Hadamard(wires = 5)
  qml.CSWAP(wires = [5,2,3])
  qml.Hadamard(wires = 5)

  return qml.state()
```

```python
#Drawing the circuit
fig,ax = qml.draw_mpl(qn_2)()
```
